```python

def plot_roc():
    
    fig = plt.figure(figsize=(15,5))
    gs = gridspec.GridSpec(2,2, )
    @figures.helpers.beautify_axis
    def plot_fs1():
        ax = fig.add_subplot(121)
        ax.plot(logReg_xy_df_FS1['1 - specificity'].values, logReg_xy_df_FS1['sensitivity'].values,
                'o-', c=colors['LogReg'], label='LogReg FS1', lw=2, markersize=6)
        ax.plot(RF_xy_df_FS1['1 - specificity'].values, RF_xy_df_FS1['sensitivity'].values,
                'o-',c=colors['RF'], label='RF FS1', lw=2, ms=6)
        ax.plot(GBC_xy_df_FS1['1 - specificity'].values, GBC_xy_df_FS1['sensitivity'].values,
                'o-',c=colors['GBC'], label='GBC FS1', lw=2, ms=6)
        ax.set_ylim([0.55,1])
        ax.set_xlim([0.0,0.5])
        return ax
    
    @figures.helpers.beautify_axis
    def plot_fs2():
        ax = fig.add_subplot(122)
        ax.plot(logReg_xy_df_FS2['1 - specificity'].values, logReg_xy_df_FS2['sensitivity'].values,
                '>--', c=colors['LogReg'], label='LogReg FS2', lw=2, ms=10)
        ax.plot(RF_xy_df_FS2['1 - specificity'].values, RF_xy_df_FS2['sensitivity'].values,
                '>--',c=colors['RF'], label='RF FS2', lw=2, ms=10)
        ax.plot(GBC_xy_df_FS2['1 - specificity'].values, GBC_xy_df_FS2['sensitivity'].values,
                '>--',c=colors['GBC'], label='GBC FS2', lw=2, ms=10)
        ax.set_ylim([0.55,1])
        ax.set_xlim([0.0,0.5])
        return ax
    
    ax1 = plot_fs1()
    ax1.spines['right'].set_visible(True)
    ax2 = plot_fs2()
    ax2.set_yticklabels([])
    
    return fig
```
