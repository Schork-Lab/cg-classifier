```python


fig = plt.figure(figsize=(15,5))
gs = gridspec.GridSpec(2,2, wspace=0.05)

@figures.helpers.beautify_axis
def plot_fs(fs, ax, fs_name='FS 1', *args, **kwargs):
    for model, model_df in fs.iteritems():
        ax.plot(model_df['1 - specificity'].values, model_df['sensitivity'].values,
                c=colors[model], label=model+' '+fs_name, *args, **kwargs)
    ax.set_ylim([0.55,1])
    ax.set_xlim([0.0,0.5])
    return ax

ax = fig.add_subplot(gs[0,0])
ax2= fig.add_subplot(gs[1,0])
plot_fs(FS1, ax, marker='o', lw=2, markersize=6)
ax.legend(loc=7)
ax.set_xticklabels([])
plot_fs(FS2, ax2, 'FS 2', marker='>', ls='--', lw=2, markersize=6)
ax2.legend(loc=7)
ax.grid(False)
ax2.grid(False)
```

```python

gs = gridspec.GridSpec(3,2)
@figures.helpers.beautify_axis
def plot_combo(fs1, fs2, model, ax):
    ax.plot(fs1[model]['1 - specificity'].values, fs1[model]['sensitivity'].values,
            'o-', c=colors[model], label=model+' FS1', lw=2, markersize=6)
    ax.plot(fs2[model]['1 - specificity'].values, fs2[model]['sensitivity'].values,
            '>--', c=colors[model], label=model+' FS2', lw=2, ms=10, alpha=0.5)
    ax.set_ylim([0.55,1])
    ax.set_xlim([0.0,0.5])
    return ax

for i, model in enumerate(FS1):
    ax = fig.add_subplot(gs[i,1])
    plot_combo(FS1, FS2, model, ax)
    if i > 0:
        ax.set_xticklabels([])
    ax.grid(False)
    ax.legend(loc=7)
    
```
