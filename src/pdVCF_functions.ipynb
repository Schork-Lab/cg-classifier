```python
# Header parsing and VCF metadata bypassing
def header_parsing_sample(var_line):
    '''This function parses the vcf header'''
    if var_line.startswith("#CHROM"):
        vcf_header_raw = var_line[1:].rstrip("\n").split("\t")
        vcf_header = vcf_header_raw[:9]
        sample_ids = vcf_header_raw[9:]
        return vcf_header, sample_ids
    
    if var_line.startswith("#"):
        return ["continue","continue"]
```

```python
def get_allele_specific_col(dataframe):
    '''  This function identifies sample genotype columns
         with allele specific information'''
    allele_specific_col = []
    for c in dataframe.columns:
        if "," in dataframe.ix[0][c]:
            allele_specific_col.append(c)
    return allele_specific_col
    
```

```python
def get_missing_example(line):
    '''  This function creates a missing sample genotype column  '''
    missing = ["./."]
    for c in line[1:]:
        if "," in c:
            missing.append(".,.")
        else:
            missing.append(".")
    return missing
    
```

```python
""" These functions create a dataframe of the sample information """

def get_sample_df(sample_data,sample_ids,format_names):
    ''' This function converts the sample data into a dataframe  '''
    
    def sample_df_filter(sample_df_copy,allele_specific_col):
        '''  Convenence function to spliting allele-specific data columns '''
        
        def col_split(sample_df,col,delim):
            ''' This function splits the col using the delimitter and 
                creates two new columns, while deleting the original col'''
            split_len = sample_df.ix[sample_df.index[0]][col].split(",")
            for n,i in enumerate(split_len):
                sample_df[col+str(n)] = sample_df[col].map(lambda x: x.split(delim)[n])
                
            return sample_df
        
        if 'CGA_CEGL' in allele_specific_col:
            allele_specific_col.pop(allele_specific_col.index('CGA_CEGL'))
            allele_specific_col.pop(allele_specific_col.index('CGA_CEHQ'))
        for c in allele_specific_col: #iterating through columns for splitting
            sample_df_copy = col_split(sample_df_copy, c, ",")
        return sample_df_copy
    
    
    def GT_split(line,gts_num):
        '''  Splitting genotype call column into allele specific columns '''
        GT = line['GT'].split(line['phase'])[gts_num]
        return GT

    
    def phase(line):
        ''' Creating phase column  '''
        phase = "/"
        if "|" in line['GT']:
            phase = "|"
        return phase
 
    
    
    #Creating dictionary to pass to DataFrame
    sample_data_df = pd.DataFrame.from_records([i.split(":") for i in sample_data])
    sample_data_df.columns = format_names
    
    #Identifying allele specific columns
    allele_specific_col = get_allele_specific_col(sample_data_df)
    
    #creating df copy
    sample_data_df_copy = sample_data_df.copy()
    
    #creating allele specific columns
    sample_data_df_copy['phase'] = sample_data_df_copy.apply(phase,axis=1)  #getting phase info
    sample_data_df_copy['GT0'] = sample_data_df_copy.apply(GT_split,axis=1,args=[0])  #first allele
    sample_data_df_copy['GT1'] = sample_data_df_copy.apply(GT_split,axis=1,args=[1])  #second allele
    
    #split allele-specific data
    sample_df = sample_df_filter(sample_data_df_copy, allele_specific_col)  #splitting data into allele-specific counts
    
    return sample_df

```

```python
#THESE FUNCTIONS NEED TO DIFFERENTIATE A CV OR SV
def __type_df(df):
    ref = df['REF']
    alt = df['ALT']
    if is_snp(ref,alt): return 'snp'
    elif is_mnp(ref,alt): return 'mnp'
    elif is_insertion(ref,alt): return 'ins'
    elif is_del(ref,alt): return 'del'
    else: return 'indel or SV'

def is_snp(ref,alt):
    """ Return whether or not the variant is a SNP """
    if len(ref) != len(alt): return False
    base_diff = [nt for i,nt in enumerate(alt) if ref[i] != alt[i]]
    if len(base_diff) <=1:
            return True
    return False


def is_mnp(ref,alt):
    """ Return whether or not the variant is an MNP"""
    if len(ref) > 1 and len(ref) == len(alt):
        return True

def is_insertion(ref,alt):
    if len(ref) < len(alt):
        return True
    return False
    
def is_del(ref,alt):
    if len(ref) > len(alt):
        return True
    return False
    
```

```python
#CONVENIENCE FUNCTION FOR NORMALIZING PADDED SNPs AND MNPs USING THE mnp_split function
def snp_normalization(df):
    df['variant_id'] = df.apply(var_id, axis=1)
    df['basechange'] = df.apply(basechange, axis=1)
    df['type'] = df.apply(__type_df, axis=1)
    df['diploid'] = df.apply(__diploid_state_df,axis=1)
    df['PS_binary'] = df.apply(phase_set, axis=1)
    
    snp_df = df[df.vartype.isin(["snp"])]
    bool_gate = [True if len(i) > 1 else False for i in snp_df['REF']]
    long_snp_df = snp_df[bool_gate]
    
    normalized_snps = mnp_split(long_snp_df)
    
    mnp_df = df[df.vartype.isin(["mnp"])]
    normalized_mnps = mnp_split(mnp_df)
    
    snp_drop = [i for i in df.index if len(df.ix[i]['REF']) > 1 and df.ix[i]['type'] == 'snp']
    #print len(snp_drop), "snps dropped from df"
    new_df = df.drop(snp_drop)
    mnp_drop = [i for i in df.index if df.ix[i]['type'] == "mnp"]
    #print len(mnp_drop), "mnps dropped from df"
    new_df = new_df.drop(mnp_drop)
    
    df = pd.concat([new_df, normalized_snps, normalized_mnps])
    
    return df


def var_id(line):
    return "_".join([line['CHROM'],line['POS'],line['REF'],line['ALT']])

def basechange(line):
    return "_".join([line['REF'], line['ALT']])

#VERY USEFUL FUNCTION FOR PULLING APART MNPs
def mnp_split(mnp_df):
    ''' The input must not contain indels variants
        as this will truncate the zip'''
    list_of_index_names = []
    series_list = []
    for i in mnp_df.index:
        variant = mnp_df.ix[i]
        start = int(variant['POS'])
        ref = variant['REF']
        alt = variant['ALT']
        assert "," not in alt
        assert len(ref) == len(alt)
        matched_bps = zip(ref, alt)
        for end,bps in enumerate(matched_bps):
            new_variant = variant.copy()
            if bps[0] != bps[1]:
                new_variant['REF'] = bps[0]  #changing ref allele to single nt
                new_variant['ALT'] = bps[1]  #changing alt allele to single nt
                new_variant['POS'] = str(start + (end))  #changing pos to single nt pos
                new_index = "_".join([new_variant['CHROM'],new_variant['POS'],bps[0],bps[1]])  #new variant_id
                new_variant['variant_id'] = new_index
                new_variant['basechange'] = bps[0]+"_"+bps[1]  #updating basechange
                series_list.append(new_variant)  #adding to list
                list_of_index_names.append(new_index)  #adding variant_id to list
    
    mnp_df_expanded = pd.DataFrame(series_list)  #creating a DataFrame
    mnp_df_expanded.index = list_of_index_names  #indexing DataFrame with new variant_ids
    return mnp_df_expanded

def phase_set(line):
    if line['PS'] == ".": return 0
    else: return 1

def __diploid_state_df(df):
    if len(df['GT']) > 1:  #autosomes
        if df['GT'][0] == "." or df['GT'][2] == ".": return "./."
        allele1, allele2 =  int(df['GT'][0]), int(df['GT'][2])
        gts_nums = [allele1, allele2]
        if allele1 == allele2 == 0:
            return '0/0'
        if allele1 != allele2 and 0 in gts_nums:
            return '1/0'
        else:
            if allele1 == allele2:
                return '1/1'
            else:
                return '1/2'
        return df
    else:  #sex chromosomes
        if df['GT'][0] == ".": return 4
        allele1 = int(df.GT[0][0])
        if allele1 == 0:
            return '0/0'
        if allele1 != 0:
            return '1/1'
        return df
```

```python
def vcf_write_str(df):
    var = list(df[['CHROM','POS','ID','REF','ALT','QUAL','FILTER','INFO','FORMAT']].astype(str))
    sample_df = list(df[var[-1].split(":")].astype(str))
    sample_df_str = ":".join(sample_df)
    var.append(sample_df_str)
    return "\t".join(var)
```

```python
def var_line_processing(var_line):
    
    parsed = var_line.rstrip("\n").split("\t",11)  #split line
    parsed_sample = parsed[:9]  #variant info no sample specific information
    format_ids = parsed[8].split(":")  #format ids
    samples_data = parsed[9:10]  #split sample data
    
    sample_df_allele = get_sample_df(samples_data,sample_ids,format_ids)  #expanding sample data inf

    variant_df_t = pd.DataFrame([i for i in parsed[:9]])  #creating df
    variant_df = variant_df_t.transpose()  #fixing orientation
    variant_df.columns = vcf_header[0:9]  #assigning columns
    variant_df = variant_df.join(sample_df_allele)  #joining the sample information
    variant_df = snp_normalization(variant_df)  #flattening the variants
    return variant_df#(parsed_sample, failing_genos, pass_alleles)
```

```python
def del_match_id(line):
    ref = line['REF'][1:]
    var_id = [str(line['CHROM']),str(line['POS']),ref,line['type']]
    return "_".join(var_id)

def del_match_id_df(df):
    df['match_id'] = df.apply(del_match_id,axis=1)
    return df



def ins_match_id(line):
    ref = line['REF'][1:]
    alt = line['ALT'][1:]
    var_id = [str(line['CHROM']),str(line['POS']),alt,line['type']]
    return "_".join(var_id)

def ins_match_id_df(df):
    df['match_id'] = df.apply(ins_match_id,axis=1)
    return df

def snp_match_id(line):
    var_id = [str(line['CHROM']),str(line['POS']),line['REF'],line['ALT'],line['type']]
    return "_".join(var_id)
    
def snp_match_id_df(df):
    df['match_id'] = df.apply(snp_match_id,axis=1)
    return df

def match_var_id(df):
    
    df_del = df[df.type=='del']
    if len(df_del) > 0:
        df_del = del_match_id_df(df_del)
    df_ins = df[df.type=='ins']
    if len(df_ins) > 0:
        df_ins = ins_match_id_df(df_ins)
    
    df_snp = df[(df.type=='snp') | (df.type=='mnp')]
    df_snp = snp_match_id_df(df_snp)
    return pd.concat([df_del, df_ins, df_snp])
```

```python
def haplotype_assign(sample,m1,m2,f1,f2):
    first = sample.map(lambda x: x.replace('/','|').split(":")[0].split('|')[0])  #parsing the first allele call from the sample of interest into an allele call vector
    second = sample.map(lambda x: x.replace('/','|').split(":")[0].split('|')[1])  #parsing the second allele call from the sample of interest into an allele call vector
    parents = [m1,m2,f1,f2]
    first_dist = [editdist.distance("".join(p),"".join(first)) for p in parents]
    second_dist = [editdist.distance("".join(p),"".join(second)) for p in parents]
    parents_assign = ['m1','m2','f1','f2']
    return parents_assign[first_dist.index(min(first_dist))], first, parents_assign[second_dist.index(min(second_dist))], second
```

```python
def phasing(genotype):
    phase = []
    for g in genotype:
        if "/" in g:
            phase.append("/")
            continue
        if "|" in g:
            phase.append("|")
            continue
        phase.append("?")
    return phase

def inheritance(genotype):
    return [i.split(":")[-1] for i in genotype]
```

```python
def cost_genotype(px, cx):
    ''' This function returns 0 penalty if child allele == expected matching 
        parental chromosome allele, else return -2 '''
    if px == cx: return 0  #matching
    return -2

def cost_recomb(px, cx):
    ''' If child allele does not match expected allele from matching parental chromosome
        AND child allele matches mis-match parental chromosome return -20 score, else
        return double penalty'''
    if px == cx: return -20
    return -22

def running_score(px, px2, cx, p1_last, p2_last):
    ''' This function returns the maximum of two running scores,
        calculated using penalties dependent on parental allele matching'''
    cont = p1_last + cost_genotype(px, cx)
    recomb = p2_last + cost_recomb(px2, cx)
    if cont > recomb:
        return cont, "Continue"
    else:
        return recomb, "Recombination"
    
for kid in kids[1:2]:  #iterate through children
    kid_chroms = get_parental_chrom(kid,df.columns)  #find column ids for specific child
    for chrom in kid_chroms:  #iterate through phased chromosomes in specific child
        p_chrom = get_chrom_obj(chrom.split("_")[-1])  #get matching parental chromosome genotypes
        alt_p_chrom = get_alt_chrom_obj(chrom.split("_")[-1])  #get mismatching parental chromosome
        
        p1_last = 0  #initialize matching parental chromosome running score 
        p2_last = 0  #initialize mismatching parental chromosome running score
        p1 = []  #matching parental chromosome score vector
        p2 = []  #mismatching parental chromosome score vector
        chrom_df = df[chrom]  #child genotype vector
        for n,loci in enumerate(chrom_df.index):  #iterate through loci
            p1_check = running_score(p_chrom[n], alt_p_chrom[n], chrom_df[loci], p1_last, p2_last)
            p1.append(p1_check)
            p1_last = p1_check[0]
            
            p2_check = running_score(alt_p_chrom[n], p_chrom[n], chrom_df[loci], p2_last, p1_last)
            p2.append(p2_check)
            p2_last = p2_check[0]
        
        

```

```python
def genotype_calls(sample,parent):
    sample_id = sample.name
    if parent == 'mother':
        return sample.map(lambda x: x.replace('/','|').split(":")[0].split('|')[0])
    if parent == 'father':
        return sample.map(lambda x: x.replace('/','|').split(":")[0].split('|')[1])

def genotype_phased(df,sample_id):
    df[sample_id+'_phase'] = phasing(df[sample_id])
    df[sample_id+'_mother'] = genotype_calls(df[sample_id], 'mother')
    df[sample_id+'_father'] = genotype_calls(df[sample_id], 'father')
    del df[sample_id]
    return df
```
